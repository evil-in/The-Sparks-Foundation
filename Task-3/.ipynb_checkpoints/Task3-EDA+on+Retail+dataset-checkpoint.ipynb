### Insights based on Conditional selection 

#### 1. The sub-category that brought in the highest sales and profit?

```python
print("The highest sales was =", data.Sales.max(), "and the highest profit was=", data.Profit.max())
```

```python
Category_max_sales = data[data.Sales == data.Sales.max()]
Category_max_sales
```

```python
Category_max_profit = data[data.Profit == data.Profit.max()]
Category_max_profit
```

from this we see that highest sales does not necessarily mean highest profit, though both in Technology category.

#### 2. The top 100 items with highest Sales:

```python
q_sales = data['Sales'].quantile(0.99)
highest_sales = data[data.Sales >q_sales]
highest_sales
```

```python
highest_sales.describe()
```
